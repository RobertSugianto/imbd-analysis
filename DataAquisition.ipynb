```python
import pandas as pd
```

```python
df = pd.read_csv('./imdb_top_1000.csv')
df.head()
```

```python
df.pop('Poster_Link')
df.head()
```

# Change columns name

```python
df.rename(columns={
    'Series_Title' : 'SeriesTitle',
    'Released_Year' : 'ReleasedYear',
    'IMDB_Rating' : 'IMDBRating',
    'Meta_score' : 'MetaScore',
    'No_of_Votes' : 'NoOfVotes'
}, inplace=True)

df.info()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df.info()
```

```python
df['Runtime'] = df['Runtime'].str.strip(' min')
df['Gross'] = df['Gross'].str.replace(',', '')
```

```python
df['ReleasedYear'].unique()
```

```python
df = df.astype({
    'SeriesTitle' : 'string',
    'Certificate' : 'string',
    'Runtime' : 'int64',
    'Genre' : 'string',
    'Overview' : 'string',
    'Director' : 'string',
    'Star1' : 'string',
    'Star2' : 'string',
    'Star3' : 'string',
    'Star4' : 'string',
    'Gross' : 'float64',
})

df.info()
```

```python
df= df[~(pd.to_numeric(df['ReleasedYear'], errors = 'coerce').isna())]
df['ReleasedYear'] = df['ReleasedYear'].astype('int64')
df.info()
```

```python
df
```

```python
df = df.dropna()
df['ReleasedYear'].unique()
```

```python
df.info()
```
